```python
import pandas
df = pandas.read_csv('Module2_RegressionData.csv')
df.head()
```

```python
# Non-Linear Regression
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# X4 data encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['X4'] = le.fit_transform(df['X4'])

# Normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pandas.DataFrame(scaler.fit_transform(df), columns=df.columns)

# Dimensionality Reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
df = pandas.DataFrame(pca.fit_transform(df), columns=df.columns)

Y = df['Y']
X = df.drop('Y', axis=1)
```

```python
# Clean data by removing outliers
from scipy import stats
z = np.abs(stats.zscore(df))
df = df[(z < 3).all(axis=1)]

# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=23)

# fit the model
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_test)
Y_train = Y_train.values.reshape(-1,1)
Y_test = Y_test.values.reshape(-1,1)

model = LinearRegression()
model.fit(X_train_poly, Y_train)

# predict
Y_pred = model.predict(X_test_poly)

# evaluate r2
from sklearn.metrics import r2_score
r2_score(Y_test, Y_pred)
```
